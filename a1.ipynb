```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
PG=wb.DataReader("^NSEI",data_source="yahoo",start="2000-1-1")
PG.head()
```

```python
PG
```

```python
PG.tail()
```

```python
PG["simple return"]=(PG["Adj Close"]/PG["Adj Close"].shift(1))-1
PG["simple return"]
```

```python
PG["simple return"].plot(figsize=(24,15))
plt.show()
```

```python
avg_returns_d=PG["simple return"].mean()
avg_returns_d
```

```python
avg_returns_a=PG["simple return"].mean()*250
avg_returns_a
```

```python
print((round(avg_returns_a,5)*100),"%")
```

```python
PG["log_return"]=np.log(PG["Adj Close"]/PG["Adj Close"].shift(1))
PG["log_return"]
```

```python
PG["log_return"].plot(figsize=(24,15))
```

```python
log_return_d=PG["log_return"].mean()
log_return_d
```

```python
log_return_a=PG["log_return"].mean()*250
log_return_a
```

```python
print(round(log_return_a,5)*100,"%")
```

```python
tickers=["^NSEI","BSE.NS"]
mydata=pd.DataFrame()
for t in tickers:
    mydata[t]=wb.DataReader(t,data_source="yahoo",start="2018-1-1")["Adj Close"]
```

```python
mydata.info()
```

```python
mydata.head()
```

```python
mydata.tail()
```

```python
(mydata/mydata.iloc[0]*100).plot(figsize=(45,18))
plt.show()
```

```python
mydata.loc["2018-01-02"]
```

```python
returns=(mydata/mydata.shift(1))-1
returns.head()
```

```python
weights=np.array([0.8, 0.2])
```

```python
np.dot(returns,weights)
```

```python
annual_returns=returns.mean()*250
annual_returns
```

```python
np.dot(annual_returns,weights)
```

```python
pfolio_1=round(np.dot(annual_returns,weights),5)*100
print(pfolio_1,"%")
```
